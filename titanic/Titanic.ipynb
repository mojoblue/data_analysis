```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
# pd.read_table('data/train.csv', sep=',')
def preprocessing(df):
    frame = df.drop(labels=['PassengerId', 'Name', 'Ticket', 'Embarked', 'Cabin'], axis=1)
    # frame = frame.replace('male', 0)
    # frame = frame.replace('female', 1)
    frame['Sex'] = frame['Sex']=='male'
    frame['Age'] = frame['Age'].replace(np.nan, int(np.average(frame['Age'].dropna())))
    f = frame
    frame = ((f-f.min(axis=0))/(f.max(axis=0)-f.min(axis=0)))
    return frame
# frame
```

```python
frame_train = preprocessing(df)
frame_test = preprocessing(df_test)

def makeTrainData(frame):
    length = len(frame)
    data, survive = [], []
    for i in range(length):
        temp = []
        for j in range(len(frame.iloc[i])):
            if j==0:
                survive.append([frame.iloc[i][j]])
            else:
                temp.append(frame.iloc[i][j])
        data.append(temp)
    data = np.array(data)
    survive = np.array(survive)
    return data, survive

def makeTestData(frame):
    length = len(frame)
    data, survive = [], []
    for i in range(length):
        temp = []
        for j in range(len(frame.iloc[i])):
            temp.append(frame.iloc[i][j])
        data.append(temp)
    data = np.array(data)
    return data

data, survive = makeTrainData(frame_train)
test_data = makeTestData(frame_test)

# print(np.array(data))
# print(np.array(sex))
```

```python
survive
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, survive, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=10)
mlp.fit(X_train, y_train)
```

```python
print("훈련 세트 정확도 : {:.2f}".format(mlp.score(X_train, y_train)))
print("테스트 세트 정확도 : {:.2f}".format(mlp.score(X_test, y_test)))
```
